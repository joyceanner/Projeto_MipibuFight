 **Lendo a planilha**

```python
import pandas as pd
sheet_id = '1gn2AoLOsYGtTpr5S0Kxl8l8_iXMg2uLBBoi72NKnFLg'
url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet'
```

```python
dados = pd.read_csv(url)
dados
```

**Identificando campos nulos e substituindo por "0"**

```python
dados.isnull().sum()
```

```python
dados = dados.fillna(0)
```

**Filtrando atletas menores de idade**

```python
atletas_menores = dados.query('IDADE < 18.0').filter(['NOME' , 'IDADE', 'EQUIPE', 'CIDADE'])
atletas_menores
```

**Salvando a lista**

```python
atletas_menores.to_excel('Menores_idade.xlsx', index=False)
```

**Filtrando atletas até 18 anos no MMA**

```python
dados['MODALIDADE'].unique()
dados = dados.replace('JIU-JITSU ', 'JIU-JITSU')
dados['MODALIDADE'].unique()
```

```python
mma = ['MMA']
lista_mma = dados.query('@mma in MODALIDADE')
```

```python
selecao_mma_18 = lista_mma['IDADE'] < 18
lista_mma[selecao_mma_18]
```

```python
selecao_mma_18.to_excel('Menores_idade_MMA.xlsx', index=False)
```
